```python
import sqlite3
import pandas as pd
```

```python
# Establish connection with internal db
con = sqlite3.connect('olist.db')
cursor = con.cursor()
```

```python
# Drop tables that already exist by running the following commands
cursor.executescript("DROP TABLE IF EXISTS customers")
cursor.executescript("DROP TABLE IF EXISTS orders")
cursor.executescript("DROP TABLE IF EXISTS order_items")
```

```python
# read csv files
customers_df = pd.read_csv("data/olist_customers_dataset.csv")
orders_df = pd.read_csv("data/olist_orders_dataset.csv")
order_items_df = pd.read_csv("data/olist_order_items_dataset.csv")
```

```python
# import data to sqlite3
customers_df.to_sql('customers', con)
orders_df.to_sql('orders', con)
order_items_df.to_sql('order_items', con)
```

```python
# verify that all data are loaded to the database
customers = pd.read_sql('SELECT * FROM customers', con)
orders = pd.read_sql('SELECT * FROM orders', con)
order_items = pd.read_sql('SELECT * FROM order_items', con)
```

```python
# We want to know for each order, what is the total price
sql1 = "SELECT order_id, SUM(price*order_item_id) AS total_value, COUNT(product_id) AS quantity FROM order_items GROUP BY order_id;"
sq1_df = pd.read_sql(sql1, con)
# sq1_df
```

```python
# See how many times each customer orders
sql3 = """
SELECT c.customer_unique_id, COUNT(o.order_id) AS order_times
FROM orders o, customers c,
(SELECT order_id, SUM(price*order_item_id) AS order_value, COUNT(product_id) AS quantity 
FROM order_items GROUP BY order_id) oi
WHERE o.order_id = oi.order_id AND c.customer_id = o.customer_id
GROUP BY c.customer_unique_id HAVING order_times > 2
ORDER BY COUNT(o.order_id) DESC
"""
```

```python
sql3_df = pd.read_sql(sql3, con)
sql3_df
```
